```python
# Install required packages
#!pip install leafmap segment-geospatial localtileserver #Uncomment and run this line to install the necessary libraries

import leafmap
from samgeo import SamGeo
import geopandas as gpd

# Initialize map centered on Kathmandu
m = leafmap.Map(center=[27.904338, 84.554139], zoom=18, height="800px")
m.add_basemap("SATELLITE")  # Add Google Satellite basemap
m
```

```python
# Display the map and instruct the user to draw an ROI
display(m)
print("Please draw a rectangle on the map to select your region of interest.")

# Get bounding box of drawn area ONLY AFTER ROI is drawn
# You may need to execute this cell again after drawing your ROI
bbox = m.user_roi_bounds()

# Download satellite imagery
image_path = "kathmandu_image.tif"
leafmap.map_tiles_to_geotiff(
    output=image_path,
    bbox=bbox,
    zoom=20,
    source="Satellite",
    overwrite=True
)

# Display downloaded image
m.add_raster(image_path, layer_name="Kathmandu Imagery")
m
```

```python
# Initialize SAM model (automatically downloads weights)
sam = SamGeo(
    model_type="vit_h",
    checkpoint="sam_vit_h_4b8939.pth",
    automatic=False,
    sam_kwargs=None
)

# Set up model with our satellite image
sam.set_image(image_path)
```
